### Stability TRP

```python
import matplotlib.pyplot as plt
import numpy as np
import itertools
import matplotlib.ticker as ticker

def intersec(list1, list2):
    intersection = len(list(set(list1).intersection(list2)))
    return float(intersection) / min(len(list1),len(list2))

def getIntersectionMatrix(usas_dict_list):
    usas_label_LL = [ list(d.keys()) for d in usas_dict_list]
    iM = np.zeros((len(usas_label_LL), len(usas_label_LL)))
    
    for i in range(0, len(usas_label_LL)):
        s1 = usas_label_LL[i]
        for j in range(0, len(usas_label_LL)):
            s2 = usas_label_LL[j]
            ii = intersec(s1,s2)
            iM[i][j] = ii

    return iM

def getUniqueLabelsCount(usas_dict_list):
    usas_label_LL = [ list(d.keys()) for d in usas_dict_list]
    ss = set([])
    for l in usas_label_LL:
        ss.update(l)
    print('>>The set contains {} different values'.format(len(ss)))
    
'''
Estimate overlap of usas labels
'''
print('trp_1')
getUniqueLabelsCount(allusas['trp_t1usas'] + [allusas['trp_t1usas_orig']] )
print('trp_2')
getUniqueLabelsCount(allusas['trp_t2usas'] + [allusas['trp_t2usas_orig']] )

'''
Estimate overlap of usas labels
'''
d2 = []
for i,d in enumerate(allusas['trp_t1usas']):
    if(i == 4):
        continue
    d2.append(d)
d2.append( allusas['trp_t1usas_orig'] )

for j,d in enumerate(allusas['trp_t2usas']):
    if(j == 3):
        continue
    d2.append(d)
d2.append( allusas['trp_t2usas_orig'] )
iM = getIntersectionMatrix(d2)
#print(iM)

'''
print the matrix
'''

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)

labels = [ r'$TRP^{W}_'+str(e)+'$' for e in range(1,5)] + [r'$TRP^{W}_{Orig}$'] + [ r'$TRP^{M}_'+str(e)+'$' for e in range(1,5)]+ [r'$TRP^{M}_{Orig}$']
#labels.append(r'$TRP^{W}_Orig$') 
#labels = ['a $e^{\sin(\omega\phi)}_{e}$','b','c','a','b','c','a','b','c','a','b','c']
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111)
cax = ax.matshow(iM, interpolation='nearest')
ax.set_yticklabels([''] + labels)
ax.set_xticklabels([''] + labels)
for tic in ax.xaxis.get_major_ticks():
    tic.tick1On = False
    tic.tick2On = True
    tic.label1On = False
    tic.label2On = True
#plt.xlabel('Predicted')
#plt.ylabel('True')

'''
Write the avg intersec between usas labels categories
'''
for i in range(len(iM)):
    for j in range(len(iM)):
        c = iM[j,i]
        ax.text(i, j, "{0:.2f}".format(c), va='center', ha='center')

ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
plt.show()
```
